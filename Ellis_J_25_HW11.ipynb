5. Write a ``ChargedParticle`` class that inherits from the ``Particle`` class, but also has an attribute for the charge of the particle.


```python
import numpy as np
class Particle(object):
    def __init__(self, mass, x, y, z, Vx, Vy, Vz):
        self.mass = mass
        self.x = x
        self.y = y
        self.z = z
        self.Vx = Vx
        self.Vy = Vy
        self.Vz = Vz
    def kinetic_energy(self):
        return 1/2 * self.mass * ((self.Vx**2)+(self.Vy**2)+(self.Vz**2))
    def distance(self, other):
        return np.sqrt(((other.x - self.x)**2) + ((other.y - self.y)**2) + ((other.z - self.z)**2))
    def new_position(self, dt):
        self.x = self.x + (dt * self.Vx)
        self.y = self.y + (dt * self.Vy)
        self.z = self.z + (dt * self.Vz)
p1= Particle(10,0,0,0,1,1,1)
p2= Particle(5, 13, 5, -3, 1, 1, 1)

print("The Kinetic Energy is " + str(p1.kinetic_energy()))        
print("The Distance between the 2 particles is " + str(p1.distance(p2)))
print(p1.x, p1.y, p1.z)
p1.new_position(10)
print(p1.x, p1.y, p1.z)


# part 5
class ChargedParticle(Particle):
    def __init__(self, mass, x, y, z, Vx, Vy, Vz, charge):
        super().__init__(mass, x, y, z, Vx, Vy, Vz)
        self.charge = charge
```

```python
import numpy as np
class Particle(object):
    def __init__(self, mass, x, y, z, Vx, Vy, Vz):
        self.mass = mass
        self.x = x
        self.y = y
        self.z = z
        self.Vx = Vx
        self.Vy = Vy
        self.Vz = Vz
    def kinetic_energy(self):
        return 1/2 * self.mass * ((self.Vx**2)+(self.Vy**2)+(self.Vz**2))
    def distance(self, other):
        return np.sqrt(((other.x - self.x)**2) + ((other.y - self.y)**2) + ((other.z - self.z)**2))
    def new_position(self, dt):
        self.x = self.x + (dt * self.Vx)
        self.y = self.y + (dt * self.Vy)
        self.z = self.z + (dt * self.Vz)
p1= Particle(10,0,0,0,1,1,1)
p2= Particle(5, 13, 5, -3, 1, 1, 1)

print("The Kinetic Energy is " + str(p1.kinetic_energy()))        
print("The Distance between the 2 particles is " + str(p1.distance(p2)))
print(p1.x, p1.y, p1.z)
p1.new_position(10)
print(p1.x, p1.y, p1.z)


class ChargedParticle(Particle):
    def __init__(self, mass, x, y, z, Vx, Vy, Vz, charge):
        super().__init__(mass, x, y, z, Vx, Vy, Vz)
        self.charge = charge
    def check_interaction(self,other):
        if self.distance(other) < 0.25:
            if self.charge * other.charge < 0:
                self.Vx, self.Vy, self.Vz = 0,0,0
                other.Vx, other.Vy, other.Vz = 0,0,0
                self.charge = 0
                other.charge = 0
                return "merge"
            elif self.distance(other) > 0:
                factor_self = (-1 * (self.charge+other.charge) * (self.mass/other.mass))
                factor_other = (-1 * (self.charge+other.charge) * (other.mass/self.mass))
                self.Vx *= factor_self
                self.Vy *= factor_self
                self.Vz *= factor_self
                other.Vx *= factor_other
                other.Vy *= factor_other
                other.Vz *= factor_other
                return "repel"
p1= ChargedParticle(5,-5,-5,-5,1,1,1,0.5)
p2= ChargedParticle(10,5, 5, 5, -1, -1, -1,-0.5)

dt= 0.05 
for step in range(300):
    p1.new_position(dt)
    p2.new_position(dt)
    result = p1.check_interaction(p2)
    print(f"Step {step:03d}:")
    print(f"  P1 pos=({p1.x:.2f}, {p1.y:.2f}, {p1.z:.2f}) vel=({p1.Vx:.2f}, {p1.Vy:.2f}, {p1.Vz:.2f}) charge={p1.charge}")
    print(f"  P2 pos=({p2.x:.2f}, {p2.y:.2f}, {p2.z:.2f}) vel=({p2.Vx:.2f}, {p2.Vy:.2f}, {p2.Vz:.2f}) charge={p2.charge}")
    if result == "merge":
        print("Interaction: Particles Merged")
        break
    elif result == "repel":
        print("Interaction: Particles Repelled")
    

p3= ChargedParticle(5,-5,-5,-5,2,2,2,0.5)
p4= ChargedParticle(10,5, 5, 5, -1, -1, -1,2)

dt= 0.05 
for step in range(300):
    p3.new_position(dt)
    p4.new_position(dt)
    result = p3.check_interaction(p4)
    print(f"Step {step:03d}:")
    print(f"  P3 pos=({p3.x:.2f}, {p3.y:.2f}, {p3.z:.2f}) vel=({p3.Vx:.2f}, {p3.Vy:.2f}, {p3.Vz:.2f}) charge={p3.charge}")
    print(f"  P4 pos=({p4.x:.2f}, {p4.y:.2f}, {p4.z:.2f}) vel=({p4.Vx:.2f}, {p4.Vy:.2f}, {p4.Vz:.2f}) charge={p4.charge}")
    if result == "merge":
        print("Interaction: Particles Merged")
        break
    elif result == "repel":
        print("Interaction: Particles Repelled")

```
